```python
#ANOVA
#The Analysis of Variance (ANOVA) is a statistical method used to test whether there are 
#significant differences between the means of two or more groups. ANOVA returns two parameters:

#F-test score: ANOVA assumes the means of all groups are the same, calculates how much the actual means deviate from the assumption, and reports it as the F-test score.
#A larger score means there is a larger difference between the means.

#P-value: P-value tells how statistically significant is our calculated score value.

#If our price variable is strongly correlated with the variable we are analyzing, expect ANOVA to return a sizeable F-test score and a small p-value.
```

```python
#Create the grouping condition:
def Job_type(value):
    if value == "Transport":
        return "Manual"
    else:
        return "Non-Manual"
```

```python
#from https://www.dataquest.io/blog/pandas-tutorial-python-2/
dforigin["Job type"]=dforigin["In which department do you work?"].apply(Job_type)
dforigin["Job type"].head()
```

```python
dforigin.head(30)
```

```python
df_group_one= dforigin[['Job type','Absences']].groupby(['Job type'],as_index=False).mean()
df_group_one.head()
```

```python
df_group_test= dforigin[['Job type', 'Absences']].groupby(['Job type'])
df_group_test.head()
```

```python
f, p = stats.f_oneway(df_group_test.get_group('Manual')['Absences'],df_group_test.get_group('Non-Manual')['Absences'])
print("ANOVA results for manual and non-manual job types: F=", round(f,2), "P=", round(p,4))
```

```python
#This is a great result, with a large F test score showing a strong correlation and 
# a P value of almost 0 implying almost certain statistical significance....
# Reference: https://www.statisticshowto.com/probability-and-statistics/f-statistic-value-test: 
#An F statistic of at least 3.95 is needed to reject the null hypothesis at an alpha level of 0.1.
#At this level, you stand a 1% chance of being wrong (Archdeacon, 1994, p.168). 
# F Values will range from 0 to an arbitrarily large number.
```
